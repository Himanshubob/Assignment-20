# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

# ans . scatter plot use the dot ploting graph 

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
```

```python
plt.xlabel("This is  x axies")
plt.ylabel("This is  y axies")
plt.scatter(x,y)
plt.show()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

#  ans..A box plot is a type of plot that shows the summary statistics of a numerical data set by using a box and whiskers diagram. It is used to visualize the distribution, shape, central value, variability, and outliers of the data

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot([box1,box2])

plt.show()
```
